# Based on certain assummptions, we will clean the data

- **Soviet Union** was dissolved in 1991, therfore, we will not consider soviet union in our model
- A total of 25 medals were awarded to the **mixed teams** in the first three Summer Olympic Games, from 1896 to 1904 and a single medal was awarded in the 1924 Winter Olympic Games.  We will not consider that.
- Similarly, we will not consider  **Russia Empire**.
- **Bohemia** is now a part of Czech Republic. So, we will remove it.
- **Czechoslovakia** broke into Czech Republic and Slovakia in 1992.
- The country **Yogoslavia**, doesn't exist anymore. We will not consider this country.
- The country of **Ceylon**, is now called Sri-Lanka.
- The **United Team of Germany** was a combined team of athletes from West Germany and East Germany that competed in the 1956, 1960 and 1964 Winter and Summer Olympic Games. Therefore, we will consider it as Germany.
- **Formosa**, is known as Taiwan. It is known as Chineese Taipei
- The region **British West Indies** is now split into different countries. We will not consider it.
- **East Germany and West Germany** unified after the fall of Berlin Wall. So, we will combine both datasets into a singular germany.
- The **Netherlands Antilles** was a group of five Caribbean islands that were part of the Kingdom of the Netherlands.Therefore, we will not consider it.
- The **Unified Team** was the name used for the sports team of the former Soviet Union (except the Baltic states) at the 1992 Winter Olympics in Albertville and the 1992 Summer Olympics in Barcelona. Therefore, we will not consider it.
- We will not consider **Independent Olympic Participants**
- **FR Yugoslavia** The Federal Republic of Yugoslavia ended in 2003. It is now called Serbia and Montenegro. Therefore, we will use Serbia and Montenegro. 
- The country of **Macedonia** is North Macedonia to be exact.
- The Refugee Olympic Team is a group made up of independent Olympic participants who are refugees. We will not consider it.
- Similarly, **Australasia** represents a broader continental region. Also, it was only during 1908 and 1912. Therefore, we will not consider it.
- Since, **Russia** is not allowed in the 2028 Olympics, we will not consider it. We will also not consider ROC.
- The medals that were distributed to the participants during **1906** were later not officially recognised by the IOC. Therefore, we will not consider 1906.
- The **1916** summer olympics were cancelled due to World War I. Therefore, we will remove all 1916 data.
- Similarly, we will also remove all data during **1940** and **1944** because of World War II.
- There is multiple mention of United Kingdom and Great Britain. We replace it with United Kingdom
- Malaya was the British Kingdom and Now called Malaysia. So, we will replace Malay with Malaysia.
- Lebanon has two short terms LIB and LBN.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
from wordcloud import WordCloud

from sklearn.preprocessing import LabelEncoder
import pickle
```

```python
url_medal_count = './data/summerOly_medal_counts.csv'
url_atheletes = './data/summerOly_athletes.csv'
url_hosts = './data/summerOly_hosts.csv'
url_programs = './data/encoded-summerOly_programs.csv'
```

```python
# Load dataset
df_medal_count = pd.read_csv(url_medal_count)

df_hosts = pd.read_csv(url_hosts)

df_programs = pd.read_csv(url_programs)

df_medal_count['NOC'] = df_medal_count['NOC'].str.strip()
```

```python
# Remove countries
countries_to_remove = ['Soviet Union','ROC','Russia','Mixed team','Russian Empire','Czechoslovakia','Bohemia',
                       'Yugoslavia','British West Indies','Netherlands Antilles',
                       'Unified Team', 'Independent Olympic Participants', 'Refugee Olympic Team', 'Australasia', 'West Germany'
                      ]

df_medal = df_medal_count[~df_medal_count['NOC'].isin(countries_to_remove)]

# Now, replace Ceylon with Sri Lanka
df_medal['NOC'] = df_medal['NOC'].str.replace('Ceylon', 'Sri Lanka')

# Now, replace East and West Germany with Germany
# df_medal['NOC'] = df_medal['NOC'].str.replace('West Germany', 'Germany')
# df_medal['NOC'] = df_medal['NOC'].str.replace('East Germany', 'Germany')

# Now, replace Formosa with Chineese Taipei
df_medal['NOC'] = df_medal['NOC'].str.replace('Formosa', 'Chinese Taipei')


# Now, replace FR Yugoslavia with Serbia and Montenegro
df_medal['NOC'] = df_medal['NOC'].str.replace('FR Yugoslavia', 'Serbia and Montenegro')

# Now, replace Great Britain with United Kingdom
df_medal['NOC'] = df_medal['NOC'].str.replace('Great Britain', 'United Kingdom')

# Now, replace Macedonia with North Macedonia
df_medal['NOC'] = df_medal['NOC'].str.replace('Macedonia', 'North Macedonia')
```

```python
# Sort by Year and reset the index
df_medal.sort_values(by='Year', inplace=True)
df_medal.reset_index(drop=True, inplace=True)
```

```python
# Now remove years that are irreleavant

# Remove years
years_to_remove = [1906,1916,1940,1944]
df_medal = df_medal[~df_medal['Year'].isin(years_to_remove)]

df_medal.to_csv('./data/cleaned_summerOly_medal_count.csv',index=False)
```

# Cleaning Athelets

- The following countries are removed in athelets dataset: ['AHO','NFL', 'YUG', 'IOA','EOR','AIN','BOH','ANZ','UNK','WIF','MAL','ROC','EUN','TCH','FRG','SAA','ROT']
- They are removed because of some problems with country beign split....
- Again, some counries code were renmaed to other.
- Now, replace Lebanon LIB to LBN
- Now, replace Vietnam , VNM with VIE
- East Germany replaced as Germany
- Now, replace Rhodesia with Zimbawe , RHO with ZIM
- Replace NBO 	 North Borneo with Malaysia NBO to MAS

```python
df_athelete = pd.read_csv(url_atheletes)
df_athelete['Event'] = df_athelete['Event'].str.strip()
df_athelete['Sport'] = df_athelete['Sport'].str.strip()
df_athelete['Medal'] = df_athelete['Medal'].str.strip()
df_athelete['Team'] = df_athelete['Team'].str.strip()
df_athelete['NOC'] = df_athelete['NOC'].str.strip()

df_athelete.drop('City', axis=1, inplace=True)


classifications_sport = df_athelete['Sport'].unique()
# Initialize and fit the label encoder
le_sport = LabelEncoder()
sport_encoded_labels = le_sport.fit(classifications_sport)

# Apply Label Encoding to the 'Category' column
df_athelete['Sport'] = le_sport.transform(df_athelete['Sport'])


#Event Encoding
classifications_event = df_athelete['Event'].unique()
# Initialize and fit the label encoder
le_event = LabelEncoder()
event_encoded_labels = le_event.fit(classifications_event)

# Apply Label Encoding to the 'Category' column
df_athelete['Event'] = le_event.transform(df_athelete['Event'])

#Medal Encoding
classifications_medal = df_athelete['Medal'].unique()

# Save the encoder to a file
with open('./pickle/sport_encoder.pkl', 'wb') as file:
    pickle.dump(le_sport, file)

with open('./pickle/event_encoder.pkl', 'wb') as file:
    pickle.dump(le_event, file)

df_athelete = pd.get_dummies(df_athelete, columns=['Medal'], prefix='Medal')
df_athelete['Medal_Gold'] = df_athelete['Medal_Gold'].apply(lambda x: 1 if x else 0)
df_athelete['Medal_Silver'] = df_athelete['Medal_Silver'].apply(lambda x: 1 if x else 0)
df_athelete['Medal_Bronze'] = df_athelete['Medal_Bronze'].apply(lambda x: 1 if x else 0)
df_athelete['Medal_No medal'] = df_athelete['Medal_No medal'].apply(lambda x: 1 if x else 0)
```

```python
# Remove certain countries
# Remove countries
codes_to_remove = ['AHO','NFL', 'YUG', 'IOA','EOR','AIN','BOH','ANZ','UNK','WIF','MAL','ROC','EUN','TCH','FRG','SAA','ROT']
df_athelete = df_athelete[~df_athelete['NOC'].isin(codes_to_remove)]

# Now, replace Lebanon
df_athelete['NOC'] = df_athelete['NOC'].str.replace('LIB', 'LBN')

# Now, replace Vietnam , VNM with VIE
df_athelete['NOC'] = df_athelete['NOC'].str.replace('VNM', 'VIE')

#East Germany replaced as Germany
df_athelete['NOC'] = df_athelete['NOC'].str.replace('GDR', 'GER')

# Now, replace Rhodesia with Zimbawe , RHO with ZIM
df_athelete['NOC'] = df_athelete['NOC'].str.replace('RHO', 'ZIM')

# Replace NBO 	 North Borneo with Malaysia
df_athelete['NOC'] = df_athelete['NOC'].str.replace('NBO', 'MAS')
```

```python
#Drop some years
years_to_remove = [1906,1916,1940,1944]
df_athelete = df_athelete[~df_athelete['Year'].isin(years_to_remove)]
```

```python
with open('./pickle/countries_mapping.pkl', 'rb') as f:
    countries_map = pickle.load(f)

def assign_country_code(value):
    return countries_map[value]
    
df_athelete['NOC'] = df_athelete['NOC'].apply(assign_country_code)
```

```python
# Sort by Year and reset the index
df_athelete.sort_values(by='Year', inplace=True)
df_athelete.reset_index(drop=True, inplace=True)

df_athelete.to_csv('./data/cleaned_summerOly_atheletes.csv',index=False)

df_athelete
```
